```python
from vampyr import vampyr1d as vp
import numpy as np
import matplotlib.pyplot as plt

world = vp.BoundingBox(corner=[-1], nboxes=[2], scaling=[2])
mra = vp.MultiResolutionAnalysis(world, 7)

# 1. What does our basis functions look like. Then we plot them...

tree_0 = vp.FunctionTree(mra)
tree_1 = vp.FunctionTree(mra)
tree_2 = vp.FunctionTree(mra)
tree_3 = vp.FunctionTree(mra)
tree_4 = vp.FunctionTree(mra)

r0 = [0.8]
beta = 1000.0
alpha = 1.0
gauss = vp.GaussFunc(coef=alpha, exp=beta, pos=r0)


vp.build_grid(tree_0, scales=0)
vp.build_grid(tree_1, scales=1)
vp.build_grid(tree_2, scales=2)
vp.build_grid(tree_3, scales=3)
vp.build_grid(tree_4, scales=4)
vp.project(-1, tree_0, gauss, max_iter=0)
vp.project(-1, tree_1, gauss, max_iter=1)
vp.project(-1, tree_2, gauss, max_iter=2)
vp.project(-1, tree_3, gauss, max_iter=3)
vp.project(-1, tree_4, gauss, max_iter=3)


df1 = tree_1-tree_0
df2 = tree_2-tree_1
df3 = tree_3-tree_2
df4 = tree_4-tree_3
```

```python
x = np.arange(0.0, 1.0, 0.001)
y = [tree_0([x]) for x in x]
plt.plot(x, y)
```
